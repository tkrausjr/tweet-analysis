## Testing the Kafka Consumer

```python
# import logging
# logging.basicConfig(level=logging.DEBUG)

from kafka import KafkaConsumer
consumer = KafkaConsumer('test', group_id=None, bootstrap_servers=['172.20.3.16'],auto_offset_reset='earliest', enable_auto_commit=False)
KafkaConsumer(auto_offset_reset='earliest', enable_auto_commit=False)
```

```python
counter = 0
for message in consumer:
    counter = counter + 1
    print(message)
    if counter == message_number:
        break
```

## More detailed message analysis

```python
counter = 0
for message in consumer:
    counter = counter + 1
    print ("Message # %s - %s:%d:%d: key=%s value=%s \n" % (str(counter), message.topic, message.partition,
                                          message.offset, message.key,
                                          message.value))
    if counter == message_number:
        break
print(type(message.value))
```
